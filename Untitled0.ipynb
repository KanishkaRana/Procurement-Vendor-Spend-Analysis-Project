```python
import  pandas as pd
```

```python
df = pd.read_csv('IT_Procurement_Vendor_Spend_Analytics.csv')
```

```python
#data inspection
df.head()
df.tail()
df.shape
df.info()
df.describe()
```

```python
df ['Purchase_Date'] = pd.to_datetime(df['Purchase_Date']) #changed purchase date datatype to date time from object
```

```python
df.duplicated().sum() #no dupliicates
```

```python
total_spend = df['Spend_Amount_INR'].sum() #total spend
print ("Total Spend",total_spend)
```

```python
top_vendors = df.groupby('Vendor_Name')['Spend_Amount_INR'].sum().sort_values(ascending=False).head(10) #top 10 vendors by spend amount
print(top_vendors)
```

```python
#spend by category
category_spend = df.groupby('Category')['Spend_Amount_INR'].sum().sort_values(ascending = False)
print(category_spend)
```

```python
#vendor performace
top_vendor_by_score = df.groupby('Vendor_Name')['Vendor_Score'].mean().sort_values(ascending=False)
print(top_vendor_by_score.head(5))
```

```python
df.to_csv("IT_Procurement_Vendor_Spend_Analytics.csv", index= False)
```
